__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Видно, что $$f_{4}(x)=f_{3}(x)-f_{1}(x)-f_{2}(x),$$ то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что данные вектора линейно зависимы.


__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

Преобразуем $f_{4}(x)$ к виду:

$$f_{4}(x)=(x+1)^{2}=x^{2}+2{x} + 1$$

Теперь можно заметить, что $$f_{4}(x)=f_{3}(x)+2f_{2}(x)+0.5f_{1}(x).$$

Следовательно, $f_{4}(x)$ является линейной комбинацией векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, а это значит, что данные вектора линейно зависимы.

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

Подмножество $L$ линейного пространства $V$ является его подпространством тогда и только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

$$1)\ u+v\in L;$$

$$2)\ \alpha\cdot u \in L.$$

а) Проведя проверку по приведенному утверждению, мы увидим, что 
 
$$(0, a, b)+(c, 0, d) = (c, a, b+d),$$ $$\alpha\cdot (0, a, b) = (0, \alpha a, \alpha b).$$
 
 
Полученный вектор $(c, a, b+d)$ не принадлежат указанному в задании множеству, то есть данное множество не является подпространством линейного пространства $\mathbb{R}^{3}$.

б) Все векторы, являющиеся линейными комбинациями векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$, являются линейным подпространством.
